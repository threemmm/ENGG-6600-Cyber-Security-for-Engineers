```python
# 12. Perform some preprocessing of the data using StandardScalar and MinMaxScalar 
# and report their difference in range and functionality.
# ---------------------------------------------------------------------------
# As we can see in the plot before and after scaling, the two features (RM and LSTAT) 
# were measured in different scales, so we needed to scale them to the same range to be able to get 
# the best result.

from sklearn import preprocessing
std_scale = preprocessing.StandardScaler().fit(boston_df[["LSTAT", "RM", "PTRATIO"]])
df_std = std_scale.transform(boston_df[["LSTAT", "RM", "PTRATIO"]])
minmax_scale = preprocessing.MinMaxScaler().fit(boston_df[["LSTAT", "RM", "PTRATIO"]]) 
df_minmax = minmax_scale.transform(boston_df[["LSTAT", "RM", "PTRATIO"]])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(boston_df['LSTAT'], boston_df['RM'], color='green', label='Input scale (Original)', alpha=0.5)
plt.scatter(df_std[:,0], df_std[:,1], color='red', label='Standardized',
alpha=0.3)
plt.scatter(df_minmax[:,0], df_minmax[:,1],
color='blue', label='min-max scaled [min=0, max=1]',
alpha=0.3)
plt.title('LSTAT and RM') 
plt.xlabel('LSTAT')
plt.ylabel('RM')
plt.legend(loc='upper left')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 13. Use a simple Linear Regression algorithm to predict the price of houses before and after scaling. 
#   (a) Split the data into 70% Training 30% Testing.
#   (b) Train and fit the model.
#   (c) Test the model and report the MAE, RMSE and R2.

# Splitting the data into 70% and 30%
from sklearn.model_selection import train_test_split
# x_features_df = boston_df[["LSTAT", "RM", "PTRATIO"]]
# y_target_df=boston_df[["MEDV"]]
x_features_df = boston_df[["LSTAT", "RM", "PTRATIO", "INDUS"]].values;
y_target_df = boston_df['MEDV'].values;
X_train, X_test, Y_train, Y_test = train_test_split(x_features_df, y_target_df, test_size = 0.3, random_state=43)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Train and fit the model without scaling.
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python

# Evaluation on training dataset
pred = model.predict(X_train);
MAE = mean_absolute_error(pred, Y_train);
MSE = np.sqrt(mean_squared_error(pred, Y_train));
r2 = r2_score(pred, Y_train);

print("Performance of the model without scaling on TRAINING dataset: \n",f"MAE: {MAE}\tMSE: {MSE}\tR2: {r2}");
```

```python
# Evaluation
pred = model.predict(X_test);
MAE = mean_absolute_error(pred, Y_test);
MSE = np.sqrt(mean_squared_error(pred, Y_test));
r2 = r2_score(pred, Y_test);

print("Performance of the model without scaling the data: \n",f"MAE: {MAE}\tMSE: {MSE}\tR2: {r2}");
```

```python
# Scaling - StandardScaler - then train and evaluate the model 
std_scale = preprocessing.StandardScaler()
X_train_std = std_scale.fit_transform(X_train)
X_test_std = std_scale.transform(X_test)

model = LinearRegression().fit(X_train_std, Y_train)
# model.fit(X_train_std, Y_train)

# Evaluation
pred = model.predict(X_test_std);
MAE = mean_absolute_error(pred, Y_test);
MSE = np.sqrt(mean_squared_error(pred, Y_test));
r2 = r2_score(pred, Y_test);

print("Performance of the model after scaling (StandardScaler): \n",f"MAE: {MAE}\tMSE: {MSE}\tR2: {r2}");
```

```python
minmax_scale = preprocessing.MinMaxScaler().fit(X_train) 
X_train_minmax = minmax_scale.transform(X_train)
X_test_minmax = minmax_scale.transform(X_test)


model = LinearRegression()
model.fit(X_train_minmax, Y_train)

# Evaluation
pred = model.predict(X_test_minmax);
MAE = mean_absolute_error(pred, Y_test);
MSE = np.sqrt(mean_squared_error(pred, Y_test));
r2 = r2_score(pred, Y_test);

print("Performance of the model after scaling (StandardScaler): \n",f"MAE: {MAE}\tMSE: {MSE}\tR2: {r2}");
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_train[:,0], X_train[:,1], color='green', label='Input scale (Original)', alpha=0.5)
plt.scatter(X_train_std[:,0], X_train_std[:,1], color='red', label='Standardized',
alpha=0.3)
plt.scatter(X_train_minmax[:,0], X_train_minmax[:,1],
color='blue', label='min-max scaled [min=0, max=1]',
alpha=0.3)
plt.title('LSTAT and RM') 
plt.xlabel('LSTAT')
plt.ylabel('RM')
plt.legend(loc='upper left')
plt.grid()
plt.tight_layout()
plt.show()
```
