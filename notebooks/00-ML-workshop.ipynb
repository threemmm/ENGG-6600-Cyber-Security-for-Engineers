### NumPy

More information: https://numpy.org/doc/stable/user/absolute_beginners.html

```python
# Import the library and set an alias for it
import numpy as np
```

```python
# Define a 1-D array with 10 zeros
zero_vector = np.zeros((10))
print(zero_vector)
```

```python
# Defne a 10-by-3 matrix of zeros
zero_matrix = np.zeros((10, 3))
print(zero_matrix)
```

```python
# Define a 1-D array with 10 ones
one_vector = np.ones((10))
print(one_vector)
```

```python
# Define a 1-D array
vector = np.array([1, 2, 3, 4, 5])
print(vector)
```

```python
# Define a 3-by-2 matrix
matrix = np.array([[1, 2], [3, 4], [5, 6]])
print(matrix)
```

```python
# Check NumPy array dimension
matrix.shape
```

```python
# Modifying the items of an array
matrix[0, 1] = 18
print(matrix)
```

#### Multiple plots

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="max", color = 'red')
plt.plot(days, min_t, label="min", color = 'blue')
plt.plot(days, avg_t, label="average", color = 'gray')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="max", color = 'red')
plt.plot(days, min_t, label="min", color = 'blue')
plt.plot(days, avg_t, label="average", color = 'gray')


plt.legend(loc='best', shadow = True, fontsize = 'medium')
```

#### Bar chart

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

#### Histogram

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) # by default number of bins is set to 10
```

```python
plt.hist(blood_sugar,rwidth=0.8,bins=4, histtype= 'bar')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend(shadow = True, fontsize = 'medium')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'], orientation='horizontal')
plt.legend(shadow = True, fontsize = 'medium')
```

#### Pie Chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=False, autopct='%1.1f%%',radius=1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=False, startangle=-45)
plt.show()
```

#### Saving the plots

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart.jpg", bbox_inches="tight", pad_inches=1, transparent=True, dpi = 600)
```
